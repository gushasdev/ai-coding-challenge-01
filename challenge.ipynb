### Using the Pandas library, work with the giver dictonary to:

- Find the average sales
- Create a vertical bar graph that shows month and value
- Create a line graph that shows month and value

### Importing Pandas Library and create the DataFrame to work with

```python
import pandas as pd
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}

df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

### Find sales average

```python
df_faturamento['valor'].mean()
```

### Create a vertical bar graph that shows month and value

```python
df_faturamento.info()
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento.plot.bar(x='data_ref', y='valor')
```

```python
df_faturamento['date_month'] = df_faturamento.data_ref.dt.month_name()
```

```python
df_faturamento.plot.bar(x='date_month', y='valor')
```

### Create a line graph that shows month and value

```python
df_faturamento.plot.line(x='date_month', y='valor')
```
